## If User wants to see the Scrapped Page simultaneous(Slow Choice)

```python
import pandas as pd 
from selenium import webdriver
from bs4 import BeautifulSoup
import time

"""def print_informations():
    for i in range(0,len(column_City)):
        print("Position---> {jname}".
              format(jname=column_Position[i]))
        print("Company ---> {jname}".
              format(jname=column_Company[i]))  
        print("City ------> {jname}".
              format(jname=column_City[i]))
        print("Link ------> {jname}".
              format(jname=column_Link[i]))
        print("UpdateDate-> {jname}".
              format(jname=column_LastUpdate[i]))
        print("\n")
"""#print_information() function used for testing

column_Position=[]
column_Company=[]
column_City=[]
column_Link=[]
column_LastUpdate=[]

translationTable = str.maketrans("ğĞıİöÖüÜşŞçÇ", "gGiIoOuUsScC")#that dictionary using for translate tr words to eng

def prepare_pagesurl(i):#prepares the url of the pages that will reached
    page_num=i
    keyword1="bilisim" 
    keyword2="telekom"
    complateurl="https://www.kariyer.net/is-ilanlari/"+keyword1+"+"+keyword2+"-"+str(page_num)+"#&jbc=3&cp="+str(page_num)
    return complateurl
    
def scrap_currentpage(soup):#fills the lists with infos
    messages=soup.find_all("div",{"class":"ilan"})
    for message in messages:
        ilan=message.find("div",{"class":"col-9"})
        if ilan.text:
            column_Position.append(ilan.find("a",{"class":"link position"}).text.translate(translationTable))
            column_Company.append(ilan.find(class_="link company").text.translate(translationTable))
            column_City.append(ilan.find("span",{"class":"city"}).text.translate(translationTable))
            column_Link.append("www.kariyer.net"+ilan.find("a",{"class":"link position"})["href"])       
            column_LastUpdate.append(message.find("p",{"class":"tarih"}).text)
              

browser=webdriver.Firefox()
browser.get(prepare_pagesurl(0))
html_doc = browser.page_source
soup = BeautifulSoup(html_doc, 'html.parser')
page_num=(soup.find("a",{"id":"lnkLastPage"}).text).split()[0] #used for take the how many page are we gonna search
#page_num=2 #for short test the code

time.sleep(1)
scrap_currentpage(soup)

if int(page_num)>=1:
    for i in range(1,int(page_num)+1):
        browser.get(prepare_pagesurl(i))
        html_doc = browser.page_source
        soup = BeautifulSoup(html_doc, 'html.parser')
        time.sleep(2)
        scrap_currentpage(soup)
    
dictionary={"Position":column_Position,"Company":column_Company,"City":column_City,"Link":column_Link,"Last Update Date":column_LastUpdate}    
datas=pd.DataFrame(dictionary)

time.sleep(5)
browser.close()

datas.to_csv (r'C:\Users\Kaan\Desktop\KariyerNetDatas2.csv', index = True, header=True)
print("!!Process has been DONE!!")
```
